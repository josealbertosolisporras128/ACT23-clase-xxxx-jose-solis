### clase: Modelo

```python
class Modelo:
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    def desfilar(self):
        return f"-> {self.nombre} está caminando con elegancia por la pasarela."

    def promover(self):
        return f"-> {self.nombre} está promocionando una nueva marca."

    def comercializar(self):
        return f"-> {self.nombre} está realizando una sesión de fotos."

def crear_modelo_interactivo(numero):
    print(f"\n--- Ingrese los datos para el Modelo {numero} ---")
    nombre = input("Nombre: ")
    telefono = input("Teléfono: ")
    edad = input("Edad: ")
    # Retornamos un nuevo objeto Modelo con los datos ingresados
    return Modelo(nombre, telefono, edad)

# --- Ejecución del programa ---

# Creamos los dos objetos interactuando con el usuario
m1 = crear_modelo_interactivo(1)
m2 = crear_modelo_interactivo(2)

# Mostramos los resultados
for m in [m1, m2]:
    print(f"\n--- Acciones de {m.nombre} (Edad: {m.edad}) ---")
    print(m.desfilar())
    print(m.promover())
    print(m.comercializar())
```
